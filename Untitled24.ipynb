```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.read_csv('DEXCAUS.csv', header = 0, index_col =0, na_values ='')
```

```python
data.index = pd.to_datetime(data.index)
data = data.ffill()
data = data['2015':'2025']
data.columns = ['exchange_rate']
data.head
```

```python
plt.figure(figsize=(12,5))
plt.plot(data['exchange_rate'])
plt.title('Exchange Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.show()
```

```python
data['log_returns'] = (data['exchange_rate'] / data['exchange_rate'].shift(1))
data.dropna(inplace=True)
```

```python
plt.figure(figsize=(12,5))
plt.plot(data['log_returns'])
plt.title('Log Returns of Exchange Rate')
plt.xlabel('Date')
plt.ylabel('Log Returns')
plt.show()
```

```python
adf_test = adfuller(data['log_returns'].dropna())

print(f'ADF Statistic: {adf_test[0]}')
print(f'p-value: {adf_test[1]}')
```

```python
fig, ax = plt.subplots(2, figsize=(12,8))

plot_acf(data['log_returns']**2, lags=40, ax=ax[0])
ax[0].set_title('ACF of Squared Log Returns')

plot_pacf(data['log_returns']**2, lags=40, ax=ax[1])
ax[1].set_title('PACF of Squared Log Returns')

plt.show()
```
