```python

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock tickers
tickers = [
    'BX',    # Blackstone Inc.
    'C',     # Citigroup Inc.
    'BLK',   # Blackrock Inc.
    'RY',    # Royal Bank of Canada
    'BRO',   # Brown and Brown Inc.
    'MET',   # Metlife Inc.
    'PYPL'   # PayPal
]

# Define the time period
start_date = '2020-10-15'
end_date = '2024-10-15'

# Download the data for the specified tickers
stock_data = {}
for ticker in tickers:
    stock_data[ticker] = yf.download(ticker, start=start_date, end=end_date)

# Download the S&P 500 data
sp500_data = yf.download('^GSPC', start=start_date, end=end_date)

# Calculate holding period returns for different periods
periods = ['6mo', '1y', '4y']
holding_period_returns = {}

for ticker in tickers:
    holding_period_returns[ticker] = {}
    for period in periods:
        if period == '6mo':
            end_date_period = pd.to_datetime('today') - pd.DateOffset(months=6)
        elif period == '1y':
            end_date_period = pd.to_datetime('today') - pd.DateOffset(years=1)
        elif period == '4y':
            end_date_period = pd.to_datetime('today') - pd.DateOffset(years=4)

        try:
            start_price = stock_data[ticker]['Adj Close'][stock_data[ticker].index >= end_date_period].iloc[0]
            end_price = stock_data[ticker]['Adj Close'].iloc[-1]
            holding_period_return = (end_price / start_price) - 1
            holding_period_returns[ticker][period] = holding_period_return
        except:
            holding_period_returns[ticker][period] = None


holding_period_returns['^GSPC'] = {}
for period in periods:
    if period == '6mo':
        end_date_period = pd.to_datetime('today') - pd.DateOffset(months=6)
    elif period == '1y':
        end_date_period = pd.to_datetime('today') - pd.DateOffset(years=1)
    elif period == '4y':
        end_date_period = pd.to_datetime('today') - pd.DateOffset(years=4)

    try:
        start_price = sp500_data['Adj Close'][sp500_data.index >= end_date_period].iloc[0]
        end_price = sp500_data['Adj Close'].iloc[-1]
        holding_period_return = (end_price / start_price) - 1
        holding_period_returns['^GSPC'][period] = holding_period_return
    except:
        holding_period_returns['^GSPC'][period] = None


# Save the results to an Excel file
df_returns = pd.DataFrame(holding_period_returns)
df_returns.to_excel('holding_period_returns.xlsx')


# Plot the returns
plt.figure(figsize=(12, 6))
for ticker in tickers + ['^GSPC']:
  plt.plot(df_returns.loc[:, ticker].index, df_returns.loc[:, ticker], label=ticker)

plt.title('Holding Period Returns')
plt.xlabel('Period')
plt.ylabel('Return')
plt.legend()
plt.show()
print(df_returns)
print(holding_period_returns)
```

```python

import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd

# Define the stock tickers and the index ticker
tickers = [
    'BX',    # Blackstone Inc.
    'C',     # Citigroup Inc.
    'BLK',   # Blackrock Inc.
    'RY',    # Royal Bank of Canada
    'BRO',   # Brown and Brown Inc.
    'MET',   # Metlife Inc.
    'PYPL'   # PayPal
]
index_ticker = '^GSPC'  # S&P 500 Index

# Define the time period
start_date = '2020-10-15'
end_date = '2024-10-15'

# Download stock prices and index prices
price_data = yf.download(tickers, start=start_date, end=end_date)
index_data = yf.download(index_ticker, start=start_date, end=end_date)

# Ensure both price_data and index_data are timezone-naive
price_data.index = price_data.index.tz_localize(None)
index_data.index = index_data.index.tz_localize(None)

# Calculate weekly returns for stocks and index
weekly_returns_stocks = price_data['Close'].resample('W').last().pct_change().dropna()
weekly_returns_index = index_data['Close'].resample('W').last().pct_change().dropna()

# Calculate monthly returns for stocks and index
monthly_returns_stocks = price_data['Close'].resample('M').last().pct_change().dropna()
monthly_returns_index = index_data['Close'].resample('M').last().pct_change().dropna()

# Initialize DataFrames to store beta values for weekly and monthly
beta_values_weekly = pd.DataFrame(index=weekly_returns_stocks.columns, columns=['Beta'])
beta_values_monthly = pd.DataFrame(index=monthly_returns_stocks.columns, columns=['Beta'])

# Calculate weekly beta for each stock
for ticker in weekly_returns_stocks.columns:
    covariance = weekly_returns_stocks[ticker].cov(weekly_returns_index)
    variance = weekly_returns_index.var()
    beta = covariance / variance
    beta_values_weekly.loc[ticker, 'Beta'] = beta

# Calculate monthly beta for each stock
for ticker in monthly_returns_stocks.columns:
    covariance = monthly_returns_stocks[ticker].cov(monthly_returns_index)
    variance = monthly_returns_index.var()
    beta = covariance / variance
    beta_values_monthly.loc[ticker, 'Beta'] = beta

# Create a scatter plot for weekly and monthly betas
plt.figure(figsize=(8, 6))
plt.scatter(beta_values_weekly['Beta'], beta_values_monthly['Beta'])
plt.xlabel('Weekly Beta')
plt.ylabel('Monthly Beta')
plt.title('Scatter Plot of Weekly vs. Monthly Beta')
plt.grid(True)

# Add labels for each stock
for ticker in beta_values_weekly.index:
  plt.annotate(ticker, (beta_values_weekly.loc[ticker, 'Beta'], beta_values_monthly.loc[ticker, 'Beta']))

plt.show()
```

```python
# Portfolio Analysis
import numpy as np
from scipy.optimize import minimize

# Define the stock tickers
tickers = [
    'BX',    # Blackstone Inc.
    'C',     # Citigroup Inc.
    'BLK',   # Blackrock Inc.
    'RY',    # Royal Bank of Canada
    'BRO',   # Brown and Brown Inc.
    'MET',   # Metlife Inc.
    'PYPL'   # PayPal
]

# Define the time period
start_date = '2020-10-15'
end_date = '2024-10-15'

# Download daily prices for the past 4 years
price_data = yf.download(tickers, start=start_date, end=end_date)

# Calculate daily returns
daily_returns = price_data['Close'].pct_change().dropna()

# Calculate the covariance matrix
covariance_matrix = daily_returns.cov()

# Define the objective function (portfolio variance)
def portfolio_variance(weights, covariance_matrix):
    return np.dot(weights.T, np.dot(covariance_matrix, weights))

# Define the constraints (weights must sum to 1)
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# Define the bounds (weights must be between 0 and 1)
bounds = [(0, 1) for _ in range(len(tickers))]

# Initial guess for weights (equal weights)
initial_weights = np.array([1 / len(tickers)] * len(tickers))

# Optimization to find minimum variance portfolio
result = minimize(portfolio_variance, initial_weights, args=(covariance_matrix),
                  method='SLSQP', bounds=bounds, constraints=constraints)

# Optimal weights for the minimum variance portfolio
optimal_weights = result.x

# Print the optimal weights
print("Optimal Weights for Minimum Variance Portfolio:")
for i in range(len(tickers)):
    print(f"{tickers[i]}: {optimal_weights[i]:.4f}")


```
