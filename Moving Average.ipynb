```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Download historical stock data
ticker = "AAPL"  # You can change this to any stock, e.g., 'TSLA', 'MSFT'
data = yf.download(ticker, start="2023-01-01", end="2024-12-31")

# Step 2: Calculate 50-day and 200-day moving averages
data["50_MA"] = data["Close"].rolling(window=50).mean()
data["200_MA"] = data["Close"].rolling(window=200).mean()

# Step 3: Plot the stock price and moving averages
plt.figure(figsize=(14, 7))
plt.plot(data["Close"], label="Closing Price", linewidth=1.5)
plt.plot(data["50_MA"], label="50-Day MA", linestyle='--')
plt.plot(data["200_MA"], label="200-Day MA", linestyle='--')
plt.title(f"{ticker} - Price with 50 & 200 Day Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```
